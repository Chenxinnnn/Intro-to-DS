# Let's look at some other data 
# NYC weather
# 2016 in Central Park
# https://www.kaggle.com/mathijs/weather-data-in-new-york-city-2016#weather_data_nyc_centralpark_2016(1).csv

```python
weather = pd.read_csv("weather.csv")
weather.head()
```

```python
weather.rename(columns = {"maximum temerature": "max_t", "minimum temperature":"min_t", 
                                             "average temperature":"ave_t"}, inplace = True)
weather.head()
```

```python
#this is for the dates
new_dates = pd.to_datetime(weather.date, format = "%d-%m-%y") #converting the strings to actual dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m')) #showing the format I want dates to appear
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=25)) #setting how often to show dates
plt.plot(new_dates,weather.max_t) #plotting dates and max temp
plt.plot(new_dates,weather.min_t) #plotting dates and min temp
plt.gcf().autofmt_xdate() #formatting datas and putting them in
plt.show() #showing the graph
```

```python
new_dates = pd.to_datetime(weather.date, format = "%d-%m-%y")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=25))
plt.scatter(new_dates,weather.max_t)
plt.scatter(new_dates,weather.min_t)
plt.gcf().autofmt_xdate()
plt.show()
```

## Correlation

```python
np.corrcoef(weather.max_t,weather.min_t)
```

### What is the $r^2$ ?

```python
np.corrcoef(weather.max_t,weather.min_t)[1][0]**2
```
